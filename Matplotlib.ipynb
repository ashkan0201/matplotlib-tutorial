```python
# Draw two graphs together

import math

data_s = []
data_c = []

for i in range(10):
    data_s.append(math.sin((i)))
    data_c.append(math.cos((i)))

plt.plot(data_s)
plt.plot(data_c)
plt.show()
```

```python
plt.plot(data_s)
plt.plot(data_c)

# Give a title to the chart
plt.title("Sin and Cos")

plt.show()
```

```python
plt.plot(data_s)
plt.plot(data_c)

plt.title("Sin and Cos")

# Giving a title to the graph x, y
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
plt.plot(data_s)
plt.plot(data_c)

plt.title("Sin and Cos")
plt.xlabel('x')
plt.ylabel('y')

# Give a guide to the chart
plt.legend(['sin(x)', 'cos(x)'])

plt.show()
```

```python
# Draw a diagram with different items
import random

# Creating two lists with the same number but random values
x = [random.randint(0,100) for i in range(10000)]
y = [random.randint(0,100) for i in range(10000)]

# This is also a list that is equal to the number of values of the other two lists. But this list is for valuation in painting.
label = [random.randint(0,100) for i in range(10000)]

# The number of this list is the same as the number of other lists, but the larger the number in this list, the larger it will be
sizes = [random.randint(0,100) for i in range(10000)]

# scatter is a function to show scattered values
# marker = the appearance of the diagram
# c = coloring valuation
# cmap = Color theme
# s = The size of the values
# alpha = transparency
plt.scatter(x, y, marker='o', c=label, cmap='plasma', s=sizes, alpha=0.5)
plt.show()
```

```python
# Using numpy to define the presentation
import numpy as np

# Creating a list of 10 members, each member has 10 members, which turns it into a 10 by 10 matrix.
checkered = []
for i in range(30):
    help_ = []
    for j in range(30):
        if (i + j) % 2 == 0:
            help_.append(1)
        else:
            help_.append(0)
    checkered.append(help_)

# Definition of presentation
data = np.array(checkered)

# Using the imshow function to draw a square histogram
plt.imshow(data, cmap='cool')

# Eliminate the x,y vector
plt.axis('off')

plt.show()
```

```python
import numpy as np

# Convert this binary code to a 13 by 13 matrix.
# It means a list with 13 list members, each list has only 13 other members
my_bin = "0000000000000001111111110001111111111100111111111110011000100011001100010001100111110111110011110001111000111111111000001010101000000110101100000011111110000000000000000"
i = 0
j = 13
splited = []

while True:
    try:
        splited.append(my_bin[i:j])
        if len(splited) == 13:
            raise
    except:
        break
    i += 13
    j += 13
finaly = []
for i in splited:
    x=[]
    for j in i:
        x.append(int(j))
    finaly.insert(0, x)
finaly.reverse()

data = np.array(finaly)

plt.imshow(data, cmap='cool')

plt.axis('off')
plt.show()
```
